```python
def determine_player_type(row):
    batting_avg = row['Batting_Average']
    bowling_avg = row['Bowling_Average']

    if batting_avg >= 20 and bowling_avg > 20:
        return 'All-rounder'
    elif batting_avg >= 25:
        return 'Batsman'
    else:
        return 'Bowler'


# Apply the function to create the 'Type' column
summary_df['Type'] = summary_df.apply(determine_player_type, axis=1)

# Print the updated DataFrame
print(summary_df)
```

```python
min_matches = 5

# Filter the DataFrame to remove players with less than min_matches matches
summary_df = summary_df[summary_df['Matches_Batted'] >= min_matches]

# Print the updated DataFrame
print(summary_df)
```

```python
summary_df['Player_Name'].nunique()
```

```python
summary_df.loc[(summary_df['Type'] == 'Batsman') & (summary_df['Stumpings'] > 0), 'Type'] = 'Wicket Keeper-Batsman'
summary_df.loc[(summary_df['Type'] == 'Bowler') & (summary_df['Stumpings'] > 0), 'Type'] = 'Wicket Keeper-Bowler'

# Print the updated DataFrame
print(summary_df)
```

```python
import json
import uuid

# Function to convert player data to JSON
def player_to_json(player_row):
    player_data = {
        'id': str(uuid.uuid4()),  # Generate a random UUID for the player ID
        'Player_Name': player_row['Player_Name'],
        'Matches_Batted': player_row['Matches_Batted'],
        'Not_Outs': player_row['Not_Outs'],
        'Runs_Scored': player_row['Runs_Scored'],
        'Highest_Score': player_row['Highest_Score'],
        'Batting_Average': player_row['Batting_Average'],
        'Balls_Faced': player_row['Balls_Faced'],
        'Batting_Strike_Rate': player_row['Batting_Strike_Rate'],
        'Centuries': player_row['Centuries'],
        'Half_Centuries': player_row['Half_Centuries'],
        'Fours': player_row['Fours'],
        'Sixes': player_row['Sixes'],
        'Catches_Taken': player_row['Catches_Taken'],
        'Stumpings': player_row['Stumpings'],
        'Matches_Bowled': player_row['Matches_Bowled'],
        'Balls_Bowled': player_row['Balls_Bowled'],
        'Runs_Conceded': player_row['Runs_Conceded'],
        'Wickets_Taken': player_row['Wickets_Taken'],
        'Bowling_Average': player_row['Bowling_Average'],
        'Economy_Rate': player_row['Economy_Rate'],
        'Bowling_Strike_Rate': player_row['Bowling_Strike_Rate'],
        'Type':player_row['Type'],
        'img_url':None
    }
    return player_data

# Convert each player's data to JSON and store in a list
players_json = [player_to_json(row) for _, row in summary_df.iterrows()]

# Convert the list of JSON objects to a JSON string
players_json_string = json.dumps(players_json, indent=4)

# Print the JSON string
print(players_json_string)
```

```python
file_path = "player_data.json"

# Open the file in write mode and save the JSON string
with open(file_path, "w") as json_file:
    json_file.write(players_json_string)

print("JSON data saved to:", file_path)
```
